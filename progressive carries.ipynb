```python
selected_columns=[]

def filter_columns(selected_columns, match_events):
    match_events=match_events[selected_columns]
    return match_events



    
```

```python
selected_columns=['pass_recipient','pass_length','location','pass_end_location','pass_outcome','player','team','minute','type','carry_end_location']
data=filter_columns(selected_columns,match_events)
```

```python
selected_team=[]
def selected_team_data(selected_team, data):
    team_data=data[data['team']==selected_team]
    return team_data
    
```

```python
selected_team='England'

team_data=selected_team_data(selected_team, data)
```

```python
team_data
```

```python
substitutions=team_data[team_data['type']=='Substitution']
substitutions=substitutions['minute']
first_sub=substitutions.min()
```

```python
first_sub
```

```python
team=team_data[team_data['minute']<first_sub]
```

```python
team
```

```python
team_passes_carries=team[team['type'].isin(['Pass', 'Carry'])] 
```

```python
team_passes_carries
```

```python
total_passes_carries=team_passes_carries['type'].value_counts()
```

```python
total_passes_carries
```

```python
# checking the number of progressive carries

import numpy as np

def carry_distance(row):
    start_x, start_y=row['location']
    end_x, end_y=row['carry_end_location']
    return np.sqrt((end_x-start_x)**2 +(end_y-start_y)**2)


team_passes_carries['carry_distance']=team_passes_carries.apply(lambda row: carry_distance(row) if row['type']=='Carry' else np.nan, axis=1)
```

```python
progressive = team_passes_carries[
    ((team_passes_carries['type'] == 'Pass') & (team_passes_carries['pass_length'] > 15)) |
    ((team_passes_carries['type'] == 'Carry') & (team_passes_carries['carry_distance'] > 10))
]
                    
```

```python
progressive_passes_carries=progressive['type'].value_counts()
```
