```python
events=sb.events(match_id=3895074)
```

```python
selected_columns=['pass_recipient','pass_length','location','pass_end_location','pass_outcome','player','team','minute','type','carry_end_location','shot_outcome', 'shot_statsbomb_xg',
       'shot_technique', 'shot_type', 'substitution_outcome']

events=events[selected_columns]
```

```python
events.head()
```

```python
team1=events[events['team']=='Bayern Munich']
team2=events[events['team']=='Bayer Leverkusen']
```

```python
team1=team1.sort_values('minute')
team2=team2.sort_values('minute')
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Group by minute and sum xG per minute
team1_minute_xg = team1.groupby('minute')['shot_statsbomb_xg'].sum().cumsum().reset_index()
team2_minute_xg = team2.groupby('minute')['shot_statsbomb_xg'].sum().cumsum().reset_index()

# Create a full minute range (0–max minute)
full_minutes = pd.DataFrame({'minute': range(0, max(events['minute'].max(), 91))})

# Merge to ensure continuity and forward-fill missing minutes
team1_plot = full_minutes.merge(team1_minute_xg, on='minute', how='left').fillna(method='ffill').fillna(0)
team2_plot = full_minutes.merge(team2_minute_xg, on='minute', how='left').fillna(method='ffill').fillna(0)

# Rename xG column
team1_plot.rename(columns={'shot_statsbomb_xg': 'cumulative_xg'}, inplace=True)
team2_plot.rename(columns={'shot_statsbomb_xg': 'cumulative_xg'}, inplace=True)

# Plot
plt.figure(figsize=(10, 6))
plt.step(team1_plot['minute'], team1_plot['cumulative_xg'], where='post', label='Bayern Munich', color='#58a6ff')
plt.step(team2_plot['minute'], team2_plot['cumulative_xg'], where='post', label='Bayer Leverkusen', color='#f97316')

plt.xlabel('Minute')
plt.ylabel('Cumulative xG')
plt.title('xG Progression Throughout the Game')
plt.grid(True, linestyle='--', alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```
