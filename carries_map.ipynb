```python

def filter_selected_columns(selected_columns,events):
    events=events[selected_columns]
    return events
```

```python
selected_columns=['pass_recipient','pass_length','location','pass_end_location','pass_outcome','player','team','minute','type','carry_end_location']
data=filter_selected_columns(selected_columns,events)
```

```python
data.head(50)
```

```python
team_selected=[]

def filter_by_team_selected(team_selected,data):
    data=data[data['team']==team_selected].reset_index(drop=True)
    return data 
```

```python
team_selected='Germany'
team_data=filter_by_team_selected(team_selected,data)
```

```python
team_data.head(50)
```

```python
# filter by the amount of carries
df=team_data

def get_carry_events(df):
    carry_events=df[df['type']=='Carry']
    return carry_events
```

```python
carries=get_carry_events(df)
```

```python
carries
```

```python
# get progressive carries

def get_progressive_carries(start, end, threshold=10):
    start_x,start_y=start
    end_x,end_y=end
    return end_x-start_x>threshold


carries['is_progressive_carry']=carries.apply(lambda x: get_progressive_carries(x['location'], x['carry_end_location']), axis=1)

progressive_carries=carries[carries['is_progressive_carry']]
```

```python
progressive_carries
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mplsoccer import Pitch , Sbopen, VerticalPitch , FontManager


background_color = '#0d1117'
line_color = '#58a6ff'
arrow_color = '#f97316'
title_color = '#58a6ff'

pitch=Pitch(pitch_color=background_color, line_color=line_color,pitch_type='statsbomb', line_zorder=2,stripe=False)
fg,ax=pitch.draw()

for idx, row in progressive_carries.iterrows():
    start_x, start_y=row['location']
    end_x, end_y=row['carry_end_location']
    pitch.arrows(start_x, start_y, end_x, end_y,width=2, headwidth=3, headlength=3, color=arrow_color, ax=ax)


ax.set_title("Progressive carries", fontsize=20, color=title_color, fontweight='bold')

fg.patch.set_facecolor(background_color)
ax.set_facecolor(background_color)
ax.spines['top'].set_color(line_color)
ax.spines['bottom'].set_color(line_color)
ax.spines['left'].set_color(line_color)
ax.spines['right'].set_color(line_color)
ax.xaxis.label.set_color(line_color)
ax.yaxis.label.set_color(line_color)
ax.tick_params(colors=line_color, which='both')
plt.show()
```

```python
# create progressive carries into the final third
def create_final_third_carries(row):
    start_x, start_y=row['location']
    end_x, end_y=row['carry_end_location']
    distance=end_x-start_x
    return distance>=10 and end_x>=80

progressive_final_third = progressive_carries[progressive_carries.apply(create_final_third_carries, axis=1)]



background_color = '#0d1117'
line_color = '#58a6ff'
arrow_color = '#f97316'
title_color = '#58a6ff'

pitch=Pitch(pitch_color=background_color, line_color=line_color,pitch_type='statsbomb', line_zorder=2,stripe=False)
fg,ax=pitch.draw()



for idx, row in progressive_final_third.iterrows():
    start_x, start_y = row['location']
    end_x, end_y = row['carry_end_location']
    pitch.arrows(start_x, start_y, end_x, end_y, width=2, headwidth=3, headlength=3, color=arrow_color, ax=ax)


ax.set_title("Progressive carries Final Third", fontsize=20, color=title_color, fontweight='bold')

fg.patch.set_facecolor(background_color)
ax.set_facecolor(background_color)
ax.spines['top'].set_color(line_color)
ax.spines['bottom'].set_color(line_color)
ax.spines['left'].set_color(line_color)
ax.spines['right'].set_color(line_color)
ax.xaxis.label.set_color(line_color)
ax.yaxis.label.set_color(line_color)
ax.tick_params(colors=line_color, which='both')
plt.show()
```
