```python
sb.matches(competition_id=9, season_id=281)
```

```python
events=sb.events(match_id=3895302)
```

```python
events.columns
```

```python
events=events[events['team']=='Bayer Leverkusen']
```

```python
selected_columns=['pass_recipient','pass_length','location','pass_end_location','pass_outcome','player','team','minute','type','carry_end_location','shot_outcome', 'shot_statsbomb_xg',
       'shot_technique', 'shot_type', 'substitution_outcome']

events=events[selected_columns]
```

```python
events.head()
```

```python
shots_taken=events[events['type']=='Shot']
```

```python
shots_taken
```

```python
from mplsoccer import Pitch
import matplotlib.pyplot as plt


pitch = Pitch(pitch_color='#0d1117', line_color='#58a6ff', pitch_type='statsbomb', line_zorder=2)
fig, ax = pitch.draw()

for idx , row in shots_taken.iterrows():
    x,y=row['location']
    shot_outcome=row['shot_outcome']

    if shot_outcome == 'Goal':
        color = '#22c55e'  # green
    elif shot_outcome == 'Blocked':
        color = '#facc15'  # yellow
    else:
        color = '#ef4444'  # red

    pitch.scatter(x, y, s=100, color=color, edgecolors='white', linewidth=1.5, ax=ax)


ax.set_title("Shots Taken", fontsize=20, color='#58a6ff', fontweight='bold')
fig.patch.set_facecolor('#0d1117')
ax.set_facecolor('#0d1117')
ax.tick_params(colors='#58a6ff')

plt.show()
```
